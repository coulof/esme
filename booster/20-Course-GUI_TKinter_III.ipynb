```python
from tkinter import *


window = Tk()
window.title('Frame widget')
window['bg']='bisque'

# create a Frame widget in the main window
Frame1 = Frame(window,borderwidth=2,relief=GROOVE)
Frame1.pack(side=LEFT,padx=10,pady=10)

# create a second Frame widget in the main window
Frame2 = Frame(window,borderwidth=2,relief=GROOVE)
Frame2.pack(side=LEFT,padx=10,pady=10)

# creating a Frame widget... within a Frame widget
# Frame1 widget is the parent of Frame3 widget
# the parent of Frame1 widget is the window widget (main window)
Frame3 = Frame(Frame1,bg="white",borderwidth=2,relief=GROOVE)
Frame3.pack(side=LEFT,padx=10,pady=10)

# creating a Label widget and a Button widget in a Frame widget
Label(Frame1,text="Dentist appointment at 3pm").pack(padx=10,pady=10)
Button(Frame1,text="Remove",fg='navy',command=Frame1.destroy).pack(padx=10,pady=10)

Label(Frame2,text="Revision for Computer Science exam").pack(padx=10,pady=10)
Button(Frame2,text="Remove",fg='navy',command=Frame2.destroy).pack(padx=10,pady=10)

Label(Frame3,text="entist appointment at 10am",bg="white").pack(padx=10,pady=10)
Button(Frame3,text="Remove",fg='navy',command=Frame3.destroy).pack(padx=10,pady=10)

window.mainloop()
```

```python
from tkinter import *
from tkinter.messagebox import * # boÃ®te de dialogue

def Verify():
    if Password.get() == 'python27':
        # le mot de passe est bon : on affiche une boÃ®te de dialogue puis on ferme la fenÃªtre
        showinfo('Result','Correct password.\nBye !')
        window.destroy()
    else:
        # le mot de passe est incorrect : on affiche une boÃ®te de dialogue
        showwarning('Result','Wrong password.\nTry again !')
        Password.set('')

# CrÃ©ation de la fenÃªtre principale (main window)
window = Tk()
window.title('Identification requise')

# CrÃ©ation d'un widget Label (texte 'Mot de passe')
Label1 = Label(window, text = 'Mot de passe ')
Label1.pack(side = LEFT, padx = 5, pady = 5)

# CrÃ©ation d'un widget Entry (champ de saisie)
Password= StringVar()
Champ = Entry(window, textvariable= Password, show='*', bg ='bisque', fg='maroon')
Champ.focus_set()
Champ.pack(side = LEFT, padx = 5, pady = 5)

# CrÃ©ation d'un widget Button (bouton Valider)
Bouton = Button(window, text ='Valider', command = Verify)
Bouton.pack(side = LEFT, padx = 5, pady = 5)

window.mainloop()
```

```python
from tkinter import *

def square():
    """ Square calculation """
    Resultat.set("Square = "+str(float(value.get())**2))

# CrÃ©ation de la fenÃªtre principale (main window)
window = Tk()
window.title("Spinbox widget")

value = StringVar()
value.set(2.0)
# CrÃ©ation d'un widget Spinbox
boite = Spinbox(window,from_=0,to=10,increment=0.5,textvariable=value,width=5,command=square)
boite.pack(padx=30,pady=10)

# CrÃ©ation d'un widget Label
Resultat = StringVar()
square()
Label(window,textvariable=Resultat).pack(padx=30,pady=10)

window.mainloop()
```

```python
from tkinter import *

def maj(newvalue):
    # nouvelle valeur en argument
    print(newvalue)
def plus():
    Valeur.set(str(int(Valeur.get())+10))
    print(Valeur.get())
def minus():
    Valeur.set(str(int(Valeur.get())-10))
    print(Valeur.get())

# CrÃ©ation de la fenÃªtre principale (main window)
window = Tk()
window.title("Scale widget")

Valeur = StringVar()
Valeur.set(50)
# CrÃ©ation d'un widget Scale
echelle = Scale(window,from_=-100,to=100,resolution=10,orient=HORIZONTAL, length=300,width=20,label="Offset",tickinterval=20,variable=Valeur,command=maj)
echelle.pack(padx=10,pady=10)

# CrÃ©ation d'un widget Button (bouton +)
Button(window,text="+",command=plus).pack(padx=10,pady=10)

# CrÃ©ation d'un widget Button (bouton -)
Button(window,text="-",command=minus).pack(padx=10,pady=10)

window.mainloop()
```

Each time the Go button is clicked, the script draws a disk with a radius of 20 pixels at a random position:

<img src=" " width="300">

```python
from tkinter import *
import random

def Circle():
    """ Draw a circle with center (x,y) and radius r """
    x = random.randint(0,Largeur)
    y = random.randint(0,Hauteur)
    r = 20
    canvas.create_oval(x-r, y-r, x+r, y+r, outline='blue', fill='blue')

def Effacer():
    """ Efface la zone graphique """
    canvas.delete(ALL)

# Create main window
window = Tk()
window.title('Circle')

# Create Canvas
Largeur = 480
Hauteur = 320
canvas = Canvas(window, width = Largeur, height =Hauteur, bg ='white')
canvas.pack(padx =5, pady =5)

# Create Button Go
BoutonGo = Button(window, text ='Go', command = Circle)
BoutonGo.pack(side = LEFT, padx = 10, pady = 10)

# Create Button Erase
ButtonErase = Button(window, text ='Erase', command = Effacer)
ButtonErase.pack(side = LEFT, padx = 5, pady = 5)

# Create Button Quitter
ButtonQuit = Button(window, text ='Quit', command = window.destroy)
ButtonQuit.pack(side = LEFT, padx = 5, pady = 5)

window.mainloop()
```

```python
from tkinter import *
import random

def Circle():
    """ Draw a circle with center (x,y) and radius r """
    x = random.randint(0,Width)
    y = random.randint(0,Height)
    r = 10
    
    # Draw a circle in the Canvas
    item = Canvas.create_oval(x-r, y-r, x+r, y+r, outline='black', fill='black')
    
    print("CrÃ©ation du cercle (item" , item ,")")

    print(Canvas.find_all())

def Undo():
    """ Efface le dernier cercle"""
    if len(Canvas.find_all()) > 1:
        item = Canvas.find_all()[-1]
        # Remove the circle
        Canvas.delete(item)
       
        print("Suppression du cercle (item" , item ,")")
        print(Canvas.find_all())

def EreaseAll():
    """ Erease all circles"""
    while len(Canvas.find_all()) > 1:
        Undo()


# CrÃ©ation de la fenÃªtre principale (main window)
Window = Tk()
Window.title('Cible')

# Image de fond
photo = PhotoImage(file="tk_cible.gif")

# CrÃ©ation d'un widget Canvas (zone graphique)
Width = 550
Height = 550
Canvas1 = Canvas(Window,width = Width, height =Height)
item = Canvas1.create_image(0,0,anchor=NW, image=photo)
print("Image de fond (item",item,")")
Canvas1.pack()

# CrÃ©ation d'un widget Button
BoutonGo = Button(Window, text ='Go', command = Circle)
BoutonGo.pack(side = LEFT, padx = 10, pady = 10)

# CrÃ©ation d'un widget Button
ButtonErease = Button(Window, text ='Erease Last shot', command = Undo)
ButtonErease.pack(side = LEFT, padx = 10, pady = 10)

# CrÃ©ation d'un widget Button
ButtonEreaseAll = Button(Window, text ='Erease All', command = EreaseAll)
ButtonEreaseAll.pack(side = LEFT, padx = 10, pady = 10)

# CrÃ©ation d'un widget Button (bouton Quitter)
ButtonQuit = Button(Window, text ='Quit', command = Window.destroy)
ButtonQuit.pack(side = LEFT, padx = 10, pady = 10)

Window.mainloop() 
```

The following script is an image browser (formats .gif .ppm .pgm), with a Menu widget :

<img src=" " width="300">

A FileDialog dialog box to search for a file :

<img src=" " width="300">

and a Canvas widget in which to display the image:

<img src=" " width="300">

```python
from tkinter import *
import tkinter.messagebox
import tkinter.filedialog

def Ouvrir():
    Canevas.delete(ALL) # on efface la zone graphique

    filename = tkinter.filedialog.askopenfilename(title="Ouvrir une image",filetypes=[('gif files','.gif'),('all files','.*')])
    print(filename)

    photo = PhotoImage(file=filename)
    gifdict[filename] = photo  # rÃ©fÃ©rence
    print(gifdict)

    Canevas.create_image(0,0,anchor=NW,image=photo)
    Canevas.config(height=photo.height(),width=photo.width())

    Mafenetre.title("Image "+str(photo.width())+" x "+str(photo.height()))

def Fermer():
    Canevas.delete(ALL)
    Mafenetre.title("Image")

def Apropos():
    tkinter.messagebox.showinfo("A propos","Tutorial Python Tkinter\n(C) Fabrice SincÃ¨re")

# Main window
Mafenetre = Tk()
Mafenetre.title("Image")

# CrÃ©ation d'un widget Menu
menubar = Menu(Mafenetre)

menufichier = Menu(menubar,tearoff=0)
menufichier.add_command(label="Ouvrir une image",command=Ouvrir)
menufichier.add_command(label="Fermer l'image",command=Fermer)
menufichier.add_command(label="Quitter",command=Mafenetre.destroy)
menubar.add_cascade(label="Fichier", menu=menufichier)

menuaide = Menu(menubar,tearoff=0)
menuaide.add_command(label="A propos",command=Apropos)
menubar.add_cascade(label="Aide", menu=menuaide)

# Affichage du menu
Mafenetre.config(menu=menubar)

# CrÃ©ation d'un widget Canvas
Canevas = Canvas(Mafenetre)
Canevas.pack(padx=5,pady=5)

# Utilisation d'un dictionnaire pour conserver une rÃ©fÃ©rence
gifdict={}

Mafenetre.mainloop()
```

1. What does the following program do?

```python
from tkinter import *

def deplacement():
    global dx, dy
    if MonCanevas.coords(balle)[3]>400:
        dy=-1*dy

    #We move the ball :
    MonCanevas.move(balle,dx,dy)
    #This function is repeated
    MaFenetre.after(20,deplacement)

#Ball movements:
dx=0
dy=5

#We create a window and a canvas:
MaFenetre = Tk()
MonCanevas = Canvas(MaFenetre,width = 500, height = 400 , bd=0, bg="white")
MonCanevas.pack(padx=10,pady=10)

#Creation of a "Quit" button:
Bouton_Quitter=Button(MaFenetre, text ='Quitter', command = MaFenetre.destroy)
#The button display is added to the tk window:
Bouton_Quitter.pack()

#We create a ball:
balle = MonCanevas.create_oval(60,10,80,30,fill='red')

deplacement()

#Start the main loop:
MaFenetre.mainloop()
```

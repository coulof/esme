    <h2>Definition</h2>
</div>

A function is a block of code. Its purpose is to execute a process and then "hand it over". It can receive data to be processed as **parameters**.

Using these parameters, the function performs an action and eventually returns a result, called the output or returned value.

A function that does not return a result is sometimes called a **procedure**.

A function can be seen as a kind of "black box":

1. To which you pass none, one or more variable(s) in brackets. These variables are called **arguments**. They can be any type of Python object.

2. To performs an action.

3. And returns a Python object or _nothing_ (i.e. `None`).

A function is introduced by the `def` keyword, followed by the name of the function, its parameters, which are enclosed in brackets (if any), a colon, and then a block of instructions that is **indented**.

```python
def my_function( parameter1 , parameter2 ,...):
    instruction block
```

La premire ligne dfinit la fonction avec les variables qui seront utilises.

Le bloc d'instructions s'appelle le **corps de la fonction**. Il doit tre obligatoirement indent, c'est--dire dire que l'on doit crer un dcalage entre ce bloc et le mot-cl **def**.

The first line defines the function and the variables that will be used.

The block of instructions is called the **function body**. It must be indented, i.e. an offset must be created between this block and the **def** keyword.

A function can perform a computation and return a value using the **return** keyword. When **return** is used, it too must be indented like the rest of the instruction block.

Any instruction written after a **return** will not be executed, so **return is said to be disconnecting**.

````python
def average(a,b):
    m = (a+b)/2
    return m

A = average(10, 6)
print(A) 

# Execute above code and see the output
````

    <h3>print vs return</h3>
</div>

The **print** instruction returns a display, but does not retrieve the result of a function. To retrieve the result of a function from outside the function, use the **return** instruction.

This **return** instruction returns the result of the calculation performed within the function.

> A beginner's mistake is to confuse the uses of `print` and `return` : 
A function with only a print and no return will only display a result on the screen **but return no value**.

Once you've defined a function, you'll need to call it to use it (line 4 of the `average` function code above). You'll also notice that the `average` function has two variables `a` and `b` as parameters.

```python
def average(a,b):
    m = (a+b)/2
    return m
A = average(10, 6)
print(A)
```

    <h2>Parameters and arguments</h2>
</div>

Une fonction peut avoir zro, un ou plusieurs paramtres.

> Although the terms **_parameters_** and **_arguments_** are often confused, there is a nuance that we'll take into account in this course: parameters are the names used in the function header, while arguments are the values passed to the function when it is called.

```python
def add(a, b):      # The parameters are a and b
    return a + b

add(5, 10)          # The arguments are 5 and 10
```

On peut passer des arguments  une fonction en utilisant les noms des paramtres, ce qui rend le code encore plus explicite. Dans ce cas l on dit que les **arguments sont nomms**.

You can pass arguments to a function using the parameter names, which makes the code even more explicit. In this case, we say that they are **named arguments**.

````python
def full_name(first_name, last_name):
    return first_name + ' ' + last_name

full_name(first_name='James', last_name='Bond')
````

Using **named arguments** allows you to pass function arguments in any order.

````python
full_name(last_name='Proust', first_name='Marcel')
````

It is possible to give **default parameters** when creating a function: the corresponding arguments are therefore no longer mandatory when calling the function. They are filled in "automatically".

````python
def full_name(first_name='John', last_name='Doe'):    # The default name is John Doe
    return first_name + ' ' + last_name

print(full_name())    # No parameters last_name et first_name are given by their default values
print(full_name('Ulysse'))    # the first_name is given, while the last_name will be by default
print(full_name(last_name='Machin'))    # The last_name is give and only the first_name will be by default
````

In the function header, we put mandatory parameters first and then default parameters. 

> Parameters with default values must always follow parameters without default values**, otherwise a syntax error will be triggered.

    <h2>Conclusion</h2>

Computer functions reach their full potential with :

 - an input, consisting of variables that serve as parameters,
 - an output (optional), which is a result eventually returned by the function (and which often depends on the input parameters).

Here's the definition of a function that calculates the sum of 3 values:

````python
def add(a, b, c):
    return a + b + c
````

We find: 

- The **def** keyword, followed by the function name and parameters followed by '**:**'.

- The **indented** function code.

- **return**, followed by the value to be returned: any instruction placed after it will not be executed.

Once the function has been defined, simply call it with the parameters of your choice:

````python
# Example 1 - simple
print(add(1,2,4))

# Example 2 - more complicated, with intermediate variables
a = 8
b = add(a, 10, 3)
print(add(5, a*2, b))
````

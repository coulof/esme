**A loop allows you to execute a piece of code several times in a row**. 

Python provides two ways of executing loops, through the `for` and `while` statements.

Although these two methods offer similar basic functionalities, their syntax and the way they check conditions differ.

<div class="alert alert-block alert-info" style='color:white; background-color:blue; border-color: blue; '>
    <h2>The loops <code>for...range</code></h2>
</div>

```
for variable in range(start, stop, step):
    instructions
```

In simple terms, `range()` allows the user to **generate a series of numbers within a given range**. You can decide where this series of numbers will begin, where it will end, and the size of the difference between one number and the next.

`range()` can take three arguments:

- `start`: integer from which the integer sequence is to be returned.
- `stop`: integer before which the integer sequence must be returned. The integer range ends at stop - 1.
- `step`: integer value that determines how much to increase to the next number (increment).

````python
for n in range(0, 4):
    print("----")
    print(n)
````

The above loop repeats the instructions 4 times, with n taking the values 0, 1, 2 and 3. 

The instruction :
```python
for n in range(0, 4):
```

indicates that we want to loop over integers from 0 (included) to 4 (excluded). We therefore stop at 3.

In the previous example, the loop starts at -2 **but does not include 3**. However, if we wish, we can use a step greater than 1 to move from one value to another...

    <p style="padding: 0 1.4rem; margin-top: .4em; margin-bottom: 0; font-size: 1em;">When you call <code>range()</code> with three arguments, you can choose not only where the series of numbers will start and stop, but also <strong>what the difference will be between one number and the next</strong>. We use the `end=` parameters to specify **what happens at the end of the print**. By default, at the end it skips the line.

````python
#... But here we end each print with a space
for i in range(3):
    print(i, end=" ")   
````

````python
#...But here we end each print with a comma and space

for k in range(3):
    print(k, end=", ")
````

By taking a negative step we can loop according to decreasing values:

````python
for l in range(0, -10, -2):
    print(l)
````

````python
for j in range(10, -1, -1):
    print(j, end="  ")
````

````python
print("Start of while instruction")
x = -2
while x < 5:
    print(x)
    x += 1 # Increase the value of x by one notch
print("End of while instruction")
````

Displays the value of `x` **as long as** x is less than 5.

<div style='color: forestgreen'>
    <h3><code>break</code> instruction </h3>
</div>

Sometimes it's useful to exit a `for` loop or a `while` loop. For example, in a `for` loop, we can check whether a condition has been met and, if so, exit the loop prematurely. For example:

````python
for x in range(10):
    print(x)
    if x == 5:
        print("It's time to get out of the loop")
        break
````

<div style='color: forestgreen'>
    <h3> <code>continue</code>instruction</h3>
</div>

Sometimes, we want to go prematurely to the next iteration of a loop, thus skipping the execution of the current iteration's instructions.

We can use `continue` to do this. Here's an example of a loop with 20 iterations (from 0 to 19) that checks whether the increment is divisible by 4.

````python
for j in range(20):
    if j % 4 == 0: # checks if j is divisible by 4
        # goes to next round (next iteration)
        continue  
    print("This value is not divisible by 4:", j)
    
````

<div style='color: forestgreen'>
    <h3><code>pass</code>Instruction </h3>
</div>

Sometimes it's useful to have an instruction that allows you to do nothing. A value in the loop is allowed to "skip its turn". For example:

````python
for x in range(-10,10):
    if x == 0:
        pass
    else:
        print(1/x)
````

This can help make the program more readable. There are specific cases where no instructions need to be executed.

Using `pass` makes it clear to anyone reading the code that the programmer's intention was to do nothing.
